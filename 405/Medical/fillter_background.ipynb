```python
import os
import cv2 as cv
import numpy as np
```

```python
rgb_dir = './dataset/train/rad/rgb'
depth_dir = './dataset/train/rad/depth'
mask_rgb_dir = './dataset/train/rad/masked_rgb'
```

```python
for folder in os.listdir(rgb_dir):
    new_rgb_dir = os.path.join(rgb_dir, folder)
    new_depth_dir = os.path.join(depth_dir, folder)
    new_mask_rgb_dir = os.path.join(mask_rgb_dir, folder)
    
    for file_name in os.listdir(new_rgb_dir):

        rgb_img_path = os.path.join(new_rgb_dir, file_name)
        depth_img_path = os.path.join(new_depth_dir, file_name)
        rgb_img = cv.imread(rgb_img_path)
        depth_img = cv.cvtColor(cv.imread(depth_img_path), cv.COLOR_BGR2GRAY)
        
        _, mask = cv.threshold(depth_img, 100, 255, cv.THRESH_BINARY)
        
        kernel = np.ones((9, 9), np.uint8)
        mask = cv.morphologyEx(mask, cv.MORPH_CLOSE, kernel)
        masked_rgb = cv.bitwise_and(rgb_img, rgb_img, mask = mask.astype('uint8'))
            
        masked_rgb_path = os.path.join(new_mask_rgb_dir, file_name)
        cv.imwrite(masked_rgb_path, masked_rgb)
```
